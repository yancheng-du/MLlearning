```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.datasets import cifar10
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
physical_devices = tf.config.list_physical_devices("GPU")
physical_devices
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
```

```python
horse_label = np.where(y_train==7)
```

```python
horse_label
```

```python
horse_img = x_train[horse_label[0]]
```

```python
x_trainlabels = {0: 'airplane',
          1: 'automobile',
          2: 'bird',
          3:'cat',
          4:'deer',
          5:'dog',
          6:'frog',
          7:'horse',
          8:'ship',
          9:'truck',}
```

```python
fig, axs = plt.subplots(2,5, figsize=(15, 6))
pic = 0
for i in range(axs.shape[0]):
    for j in range(axs.shape[1]):
        axs[i,j].imshow(horse_img[pic])
        #axs[i,j].set_title(labels[y_train[pic][0]])
        pic+=1
```

```python
model = keras.Sequential(
    [layers.InputLayer(input_shape = (32,32,3)),
     
     layers.Conv2D(32, (3,3), padding='same',activation='relu', kernel_regularizer = keras.regularizers.l2(l=0.01)),
     layers.MaxPooling2D(pool_size = (2,2)),
     layers.BatchNormalization(),
     
     layers.Dropout(0.2),
     layers.Conv2D(64, (3,3), padding='same',activation='relu', kernel_regularizer = keras.regularizers.l2(l=0.01)),
     layers.MaxPooling2D(pool_size = (2,2)),
     layers.BatchNormalization(),
     
     layers.Dropout(0.2),
     layers.Conv2D(128, (3,3), padding='same',activation='relu'),
     
     layers.Flatten(),
     
     layers.Dense(64, activation = 'relu', kernel_regularizer=keras.regularizers.l2(l = 0.01),),
     layers.Dense(32, activation = 'relu', kernel_regularizer=keras.regularizers.l2(l = 0.01),),
     layers.Dense(10)
    ]
)
```

```python
model.summary()
```

```python
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['accuracy'],
    optimizer = keras.optimizers.Adam()
)
```

```python
model.fit(x_train,y_train,
          batch_size=64,
          epochs = 5,
          verbose = 2
         )
```

```python
model.evaluate(x_test, y_test,)
```
